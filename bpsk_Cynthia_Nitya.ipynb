```python
from scipy.io import wavfile
# import bpsk
import numpy as np
import matplotlib.pyplot as mplib
import thinkdsp
from array import array

# bpsk.help()
```

```python
#FUNCTIONS GIVEN TO US

def string2NPArray(s):
    bits = np.array([])
    for a in bytearray(s, 'ascii'):
        for b in range(0,7):
            bits = np.append(bits,float((a>>(7-b-1))&1))
    return bits


def convert_to_int16(sig):
    # convert into int16  to write as wave
    sig = (sig/np.max(sig))*(2**14)
    sig = sig.astype('int16')
    return sig


def find_start_and_end(xf, threshold = 2000):   
    start_idx = -1
 
    for k in range(0, len(xf)):
        if(np.abs(xf[k])) > threshold:
            start_idx = k
            break

    if(start_idx  < 0):
        print "Unable to detect start of transmission"
        return -1
    
    end_idx = -1
    
    for k in range(0, len(xf)):
        if(np.abs(xf[len(xf)-k-1])) > threshold:
            end_idx = len(xf)-k-1
            break

    if(end_idx < 0):
        print "Unable to detect end of transmission"
        return -1

    return start_idx, end_idx

def NPbits2String(bits):
    S = ""
    for a in np.arange(0, np.ceil(len(bits)/7)):
        tmp = 0
        for k in np.arange(0,7):
            b = bits[a*7+k]
            tmp = tmp + (2**(6-k))*b
        S = S + chr(int(tmp))
    return S
```

## bpsk.generate_bpsk_signal

```python
#Takes 1s and 0s from np array and returns a new array of 250 1s for every 1 and 250 zeros for every 0. 
#Also appends 250 0s at the beginning, 250 1s right after (to look for positive signal) and 250 0s at the end.
def extend_250(bits):
    new = []
    new.extend(250*[0])
    new.extend(250*[1.0])
    for i in bits:
        if i == 1:
            new.extend(np.asarray(250*[1.0]))
        else:
            new.extend(np.asarray(250*[-1.0]))
    new.extend(np.asarray(250*[0]))
    return np.asarray(new)
```

```python
hello = string2NPArray('Hello')

#Extended_byte is the binary representation of the string, extended to make sure that each symbol length is 250
extended_byte = extend_250(hello)
rate=8820
t = np.arange(0, len(extended_byte)/float(rate), 1/float(rate))
f_c = 1000 #Frequency of the carrer-wave

# In order to find the signal of the binary representation and apply modulation, we multiply the byte by cosine
#The discontinuities in the cosine signal indicates the flipping and change in the bit
sig = extended_byte*7500.0*np.cos(2.0*np.pi*f_c*t)
        
mplib.plot(sig)
```

```python
#Writes the wave of the signal with frequency rate (Hz), in order to be played
wavfile.write('CyNitya.wav',8820,convert_to_int16(sig))
```

## bpsk.decode_bpsk_signal

```python
#Read wave and plot
fs, x = wavfile.read('AcouTestAud.wav')
thinkdsp.Wave(x, framerate=8820).plot()
```

```python
#finds starting and ending index of the signal and sets them as x
start,end=find_start_and_end(x, threshold =10000) #THIS THRESHOLD NEEDS TO BE 5000 FOR THE TEST WAVE AcousticModemRxTest.wav
mplib.plot(x[start:end])
x=x[start:end]
```

This is only the signal itself.

```python
rate = fs
freq = 1000 #frequency of cosine wave
```

```python
#Multiplies the signal by a cosine

t = np.arange(0, len(x)/float(rate), 1/float(rate)) #make t the right length spaced by the rate
phis=np.arange(-np.pi,np.pi,np.pi/6.0) #phase offset array
max_w_phi = 0
for i in range(len(phis)):
    num = max(x*np.cos(2*np.pi*freq*t+phis[i])) #find the phi to make x*cos the maximum value
    if num > max_w_phi:
        max_w_phi = num
        phi = phis[i] #set the phi to make x*cos the max value

sig_decode=x*np.cos(2*np.pi*freq*t+phi)#multiply the signal by the cosine offset properly to make x*cos the max value
mplib.plot(sig_decode)
```

```python
#take sig_decode to frequency domain with fft. Need to shift to 'center' it.
spectrum_sig = np.fft.fftshift(np.fft.fft(sig_decode))
mplib.plot(spectrum_sig)
```

We want just that middle chunk, so we need to put this signal through a low pass filter. We chose in the next step to do this by convolving sig_decode with a sinc function.

```python
#Convolving signal with a sinc with the frequency we want to cutoff at. 
f_sinc=150 #the cutoff frequency
t_sinc = np.arange(0, len(sig_decode)/float(rate), 1/float(rate))
h = np.sinc(2*t_sinc*f_sinc)
convolved = np.convolve(sig_decode,h)
spectrum_con = np.fft.fftshift(np.fft.fft(convolved))
mplib.plot(convolved)
#mplib.plot(spectrum_con) #just to verify the low pass filter
```

This has all the 1s and 0s we want to look at! 

```python
#Figures out whether the signal is inverted or not and turns it into not being inverted. 
#Also turns >0 into 1 and <0 into 0.

j = []

inverted=False
if convolved[10]<0:
    inverted=True

    
for i in convolved:
    if i > 0 and not(inverted):
        j.append(1)    
    if i<0 and inverted:
        j.append(1)
    if i>0 and inverted:
        j.append(0)
    if i<0 and not(inverted):
        j.append(0)
mplib.plot(j)
```

```python
#takes groups of 250 bits and turns them into 1 bit based on if there are mostly 1s or zeros in the region of 250 bits
l = []
for i in range(len(j)/250):
    region = j[250*i:249+250*i] #section of 250 bits
    sum_region = sum(region)
    el=float(float(sum_region)/float(len(region))) #average of the region
    
    if el>0.9: #if mostly 1s, the average will be more than 0.9
        l.append(1)
    if el<0.1: #if mostly 0s, the average will be less than 0.1
        l.append(0)
decoded_bits = np.asarray(l[1:]) #shave off that first bit from l because 
#it is just a header that we put in the generated signal 
```
